```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn import linear_model
import statsmodels.api as sm

df = pd.read_csv('Housing_group(1).csv')
df.info() # take a look at what is in the data set 
```

```python
# Do some basic explorative data analysis to make sure the outcomes from the more complicated analysis is inline with our consensus.


# print('Sum of populations per ocean proximity', df.groupby('ocean_proximity').population.sum(), '\n')
# print('Sum of populations per ocean proximity', df.groupby('ocean_proximity').households.sum(), '\n')
# print('Sum of populations per ocean proximity', df.groupby('ocean_proximity').housing_median_age.sum(), '\n')

# inc_per_prox = df.groupby('ocean_proximity').median_income.mean()

# for x in inc_per_prox:
#     print(type(x))

# print(inc_per_prox)

df = df.dropna()

x = df[['housing_median_age', 'total_rooms', 'households', 'population', 'total_bedrooms']]
y = df['median_income']


regr = linear_model.LinearRegression()
regr.fit(x, y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

# with statsmodels
x = sm.add_constant(x) # adding a constant

model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)


# print('Mean of median income per ocean proximity', df.groupby('ocean_proximity').median_income.mean(), '\n')
```
